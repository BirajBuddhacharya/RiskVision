```python
import os; os.chdir('../..') # changing to root directory of project
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pickle # saving trained model
```

```python
# saving train data
X_train = pd.read_csv('data/model_training/heart_disease/X_train.csv')
X_test = pd.read_csv('data/model_training/heart_disease/X_test.csv')
y_train = pd.read_csv('data/model_training/heart_disease/y_train.csv')
y_test = pd.read_csv('data/model_training/heart_disease/y_test.csv')
```

```python
# training model 
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict_proba(X_test)[:, 1]
```

```python
# model evaluation 
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
# saving model 
with open('models/heart_disease/heart_disease.pkl', 'wb') as fp: 
    pickle.dump(model,fp)
```
